```python
num_workers = 48 #can't always use number of physical cores because of memory constraints
```

# Grid Search

```python
out_dir = "grid_search"
```

```python
topN_search_exp = Experiment.load_from_json(
    file_dir = out_dir,
    file_name = "keyfile.json",
    out_dir = out_dir,
    fullscan_dir = multibeer_dir,
    amend_result_path = True
)
```

```python
isolation_width = 1.0
topN_search_exp.evaluate(
    num_workers = num_workers,
    isolation_widths = isolation_width,
    aligned_names = "multibeers" + template_suffix,
    mzmine_templates = mzmine_template,
    mzmine_exe = mzmine_path,
    force_peak_picking = False
)

topN_search_exp.summarise(num_workers=num_workers, rank_key="cumulative_intensity_proportion")
```
